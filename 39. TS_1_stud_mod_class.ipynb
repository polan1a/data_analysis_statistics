```python
def fill_n_mean_back(df, name, n_lag_input):
  df_in = df.copy()
  df_in['INDICE'] = np.arange(len(df)) + 1
  indices_nulos = df_in.loc[df_in[name].isnull(),'INDICE'].values
  df_in[f'{name}_input'] = df_in[name].copy()
  for i in indices_nulos:
    mask_loc = (df_in['INDICE']<i) & (df_in['INDICE']>(i -n_lag_input-1))
    df_in.loc[(df_in['INDICE']==i), f'{name}_input'] = df_in.loc[mask_loc, f'{name}_input'].mean()
  return df_in
```

# Ruido Blanco

$X_t$ ~ $N(0,1)$ con $t\in[0,1]$

```python
df_simulado = pd.DataFrame()
df_simulado['t'] = np.linspace(0,1, 1000)
df_simulado['X'] = [np.random.normal(loc=0, scale=1, size=1)[0] for x in df_simulado['t']]
px.line(df_simulado, x='t', y='X')
```

# Proceso con cambio en la media pero sin cambio en la varianza del ruido
$X_t$ ~ $N(f(t),1)$ con $t\in[0,1]$

$X_t = f(t) + \epsilon _t$ con $\epsilon _t  ~ $N(f(t),1)$

```python
df_simulado = pd.DataFrame()
df_simulado['t'] = np.linspace(0,1, 1000)
df_simulado['X_1'] = [np.random.normal(loc=100*x, scale=1, size=1)[0] for x in df_simulado['t']]
df_simulado['X_2'] = [100*x + np.random.normal(loc=0, scale=1, size=1)[0] for x in df_simulado['t']]
px.line(df_simulado, x='t', y=['X_1', 'X_2'])
```

```python
x = np.linspace(0, np.pi*10, 100)
y = np.cos(x)
data_simulada_1 = pd.DataFrame()
data_simulada_1['X'] = x
data_simulada_1['SIN_TENDENCIA'] = y
data_simulada_1['SIN_TENDENCIA_CON_RUIDO'] =  y + np.random.normal(loc=0, scale=0.5, size=100)
data_simulada_1['TENDENCIA_ADITIVA'] = y + np.exp(x/10)
data_simulada_1['TENDENCIA_ADITIVA_CON_RUIDO'] = y + np.exp(x/10) + np.random.normal(loc=0, scale=0.6, size=100)
data_simulada_1['TENDENCIA_MULTIPLICATIVA'] = y*np.exp(x/10)
data_simulada_1['TENDENCIA_MULTIPLICATIVA_CON_RUIDO'] = y*np.exp(x/10) +np.random.normal(loc=0, scale=3, size=100)
px.line(data_simulada_1, x='X', y= ['SIN_TENDENCIA','SIN_TENDENCIA_CON_RUIDO'])
```

# Proceso con cambio en la varianza del proceso
$X_t$ ~ $N(f(t),1)$ con $t\in[0,1]$

$X_t = f(t) + \epsilon _t$ con $\epsilon _t  ~ $N(f(t),1)$

```python
df_simulado = pd.DataFrame()
df_simulado['t'] = np.linspace(0,1, 1000)
df_simulado['X_1'] = [np.random.normal(loc=0, scale=x, size=1)[0] for x in df_simulado['t']]
df_simulado['X_2'] = [np.random.normal(loc=0, scale=1, size=1)[0] for x in df_simulado['t']]
px.line(df_simulado, x='t', y=['X_1', 'X_2'])
```

# Caminatas aleatorias

```python
n_pasos = 2000
punto_inicial = 0
x_lista_caminata = [punto_inicial]
for i in range(n_pasos):
  val_agregar = x_lista_caminata[-1] + np.random.normal(loc=0, scale=1, size=1)[0] # X_t = X_{t-1} + e
  x_lista_caminata.append(val_agregar)
px.line(np.array(x_lista_caminata))
```
