```python
import pandas as pd
df_practica = pd.read_csv('DATA_GEO_CUT.csv')[['LATITUD', 'LONGITUD']]
```

```python
df_practica
```

**OJO no se deberían medir distancias sólo lo hacemos como ejercicio de clusters, pero lo correcto es pasar a coordenadas cartesianas y luego ejecutar el análisis**

```python
import plotly.express as px
```

```python
df_practica_sample = df_practica.sample(5_000)
```

```python
px.scatter(df_practica_sample, x='LONGITUD', y='LATITUD')
```

```python
import sklearn.cluster as sk_clu
```

```python
tec_cluster = sk_clu.KMeans(n_clusters=20)

tec_cluster.fit(df_practica_sample[['LATITUD', 'LONGITUD']])
df_practica_sample['LABELS'] = tec_cluster.labels_
px.scatter(df_practica_sample, x='LONGITUD', y='LATITUD', color='LABELS')
```

**SUponga que una intervención afecta a todos los puntos en un determinado radio**

```python
tec_cluster = sk_clu.MeanShift(bandwidth =0.2)

tec_cluster.fit(df_practica_sample[['LATITUD', 'LONGITUD']])
df_practica_sample['LABELS'] = tec_cluster.labels_
px.scatter(df_practica_sample, x='LONGITUD', y='LATITUD', color='LABELS')
```

**Elementos conectados por una distancia menor a un valor fijo pertenezcan al mismo grupo, podemos utilizar dbscan**

```python
tec_cluster = sk_clu.DBSCAN(eps=0.2)

tec_cluster.fit(df_practica_sample[['LATITUD', 'LONGITUD']])
df_practica_sample['LABELS'] = tec_cluster.labels_
px.scatter(df_practica_sample, x='LONGITUD', y='LATITUD', color='LABELS')
```

```python
tec_cluster = sk_clu.DBSCAN(eps=0.5)

tec_cluster.fit(df_practica_sample[['LATITUD', 'LONGITUD']])
df_practica_sample['LABELS'] = tec_cluster.labels_
px.scatter(df_practica_sample, x='LONGITUD', y='LATITUD', color='LABELS')
```
